## Versuch 212 - Viskosität

```python
# Importieren der Module
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
#Importieren der Daten
r2, dr2, Lambda, vmitkor, Fehler =np.loadtxt('/Users/Peter/Desktop/Python-Kurs/Versuch212.txt', skiprows=1, usecols=(0,1,2,3,4), unpack=True)
```

```python
#Fitten und Parameter ausgeben, Ursprungsgerade
def gerade(x, A): 
    return A*x

r20=r2[0:4:1]
vmitkor0=vmitkor[0:4:1]
popt, pcov = curve_fit(gerade, r20, vmitkor0)
print(popt)
perr = np.sqrt(np.diag(pcov))
print(perr)
```

```python
#Plotten
plt.errorbar(r2, vmitkor, linestyle='None', marker='x', label='$Messwerte$', yerr=Fehler)
plt.plot(r2, gerade(r2, *popt))
plt.xlabel('Quadrierter Radius $r^2 [cm^2]$')
plt.ylabel('v/(deltarho*lambda) $10^{-3} m^4/(kg*s)$')
plt.title('Diagramm 1: Korr. Geschwindigkeit durch Dichtedifferenz über quadriertem Radius')
plt.legend(loc='best')
plt.savefig('212 Diagramm 1.png', dpi=300)
```
