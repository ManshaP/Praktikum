```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
f, df, Uss, dUss =np.loadtxt('/Users/Peter/Desktop/Python-Kurs/245Tab1.txt', skiprows=1, usecols=(0,1,2,3), unpack=True)
```

```python
# Definition der Ausgleichsgerade y=f(x)
def linear(x, A, B): 
    return A*x + B


popt, pcov = curve_fit(linear, f, Uss)
print (popt, pcov)
perr = np.sqrt(np.diag(pcov))
```

```python
plt.errorbar(f, Uss, linestyle='None', marker='>', label='$Messung, I_{HH}=4  A $', xerr=df, yerr=dUss)
plt.plot(f, linear(f, *popt), label='$m_{fit}=%.2f \pm%.2f, (t_{fit}=%.2f \pm%.3f)$' %(popt[0], perr[0], popt[1], perr[1]))
plt.xlabel('Frequenz  $f [Hz]$')
plt.ylabel('Induktionsspannung  $U_{SS} [V]$')
plt.title('Diagramm 1: Induktionspannung als Fkt. der Frequenz $U_{SS}=m*f+t$')
plt.legend(loc='best')
plt.savefig('245 Diagramm 1.png', dpi=300)
```
